# Xbox Game Pass

My intent is to provide a better understanding of what is included in the Xbox Game Pass (XGP) subscription. I plan to cull stats on the service's content, to provide insight into what is offered and what value a subscriber can expect.

### Compiling the list of games
Scraping the entire list of games through BeautifulSoup and Python proved difficult given how the Xbox Game Pass website was structured. Instead I opted to just save the webpage as two seperate HTML text documents and then parse relevant information from those local files.

```python
# create dataframe
!pip install pandas
import pandas as pd

columns = ['GameTitle', 'AvailableOn']
df_games_list = pd.DataFrame(columns = columns)
```

```python
# read in html and parse info
import requests
!pip install bs4
from bs4 import BeautifulSoup

columns = ['ProductName', 'Platform', 'ListPrice', 'MSProduct', 'Multiplayer', 'Rating', 'RatingSystem', 'ReleaseDate', 'MSStoreLink']
df_game_list = pd.DataFrame(columns = columns)
for i in [1, 2]:
    file_name = 'Xbox Game Pass {}.txt'.format(i)
    soup = BeautifulSoup(open(file_name, 'rb'), 'html.parser')
    games = soup.find_all(lambda tag: tag.name == 'section' and tag.get('class') == ['m-product-placement-item', 'f-size-medium', 'context-game', 'gameDiv'])
    for game in games:
        list_price = game.get('data-listprice')
        ms_product = game.get('data-msproduct')
        multiplayer = game.get('data-multiplayer')
        rating = game.get('data-rating')
        rating_system = game.get('data-ratingsystem')
        release_date = game.get('data-releasedate')[:game.get('data-releasedate').find('T')]
        ms_store_link = game.find(lambda tag: tag.name == 'a' and tag.get('class') == ['gameDivLink']).get('href')
        product_name = game.find(lambda tag: tag.name == 'h3' and tag.get('class') == ['c-heading']).text
        if len(game.find_all(lambda tag: tag.name == 'div' and tag.get('class') == ['c-tag'])) > 1:
            platform = 'Both'
        else:
            platform = game.find(lambda tag: tag.name == 'div' and tag.get('class') == ['c-tag']).text
        row = df_game_list.shape[0]
        df_game_list.loc[row, :] = [product_name, platform, list_price, ms_product, multiplayer, rating, rating_system, release_date, ms_store_link]
df_game_list.head()
```

```python
# saving dataframe as csv
df_game_list.to_csv('df_game_list.csv')
```

Closer inspection of the csv file reveals some formatting issues with the parsed data, namely inconsistent title schemas and character conversion errors. Corrections were applied manually and then imported as 'df_game_list_clean'.

```python
# importing 'df_games_list_clean'
df_game_list_clean = pd.read_csv('df_game_list_clean.csv').drop(columns = 'Unnamed: 0')
df_game_list_clean.head()
```

### Platform availability breakdown

All games will be separated into one of three groups: 'Xbox One', 'PC', or 'Both'. These represent games that are available on Xbox only, PC only, or both platforms, respectively. The Xbox Game Pass subscription is avilable in three flavors, PC games only for 10USD a month, Xbox games only for 10USD a month, or the entire catalog for 15USD a month. It should be interesting to see how the entire list of games splits based on these offerings.

```python
# platform availability breakdown
duplicate_count = int(df_game_list_clean[df_game_list_clean.duplicated('ProductName', keep = False)].shape[0]/2)
xbox_count = df_game_list_clean[(df_game_list_clean.duplicated('ProductName', keep = False) == False) & (df_game_list_clean.Platform == 'Xbox One')].shape[0]
pc_count = df_game_list_clean[(df_game_list_clean.duplicated('ProductName', keep = False) == False) & (df_game_list_clean.Platform == 'PC')].shape[0]
both_count = df_game_list_clean[(df_game_list_clean.duplicated('ProductName', keep = False) == False) & (df_game_list_clean.Platform == 'Both')].shape[0]
total_count = duplicate_count + xbox_count + pc_count + both_count
print(r'BREAKDOWN NUMBERS (% of total, count)')
print('  Single platform: {}, {}'.format('{:.0%}'.format((xbox_count + pc_count)/total_count), xbox_count + pc_count))
print('    Xbox only: {}, {}'.format('{:.0%}'.format(xbox_count/total_count), xbox_count))
print('    PC only: {}, {}'.format('{:.0%}'.format(pc_count/total_count), pc_count))
print('  Dual platform: {}, {}'.format('{:.0%}'.format((both_count + duplicate_count)/total_count), both_count + duplicate_count))
print('Total: {}, {}'.format('100%', total_count))
```

```python
# basic visualization
!pip install numpy
import numpy as np
!pip install matplotlib
import matplotlib.pyplot as plt

labels = ['Single', 'Dual', 'Total']
xbox = [xbox_count, 0, 0]
pc = [pc_count, 0, 0]
both = [0, both_count + duplicate_count, 0]
total = [0, 0, total_count]
width = .4
fig, ax = plt.subplots(figsize = (20, 10))
ax.bar(labels, xbox, width, bottom = pc, label = 'Xbox only')
ax.bar(labels, pc, width, label = 'PC only')
ax.bar(labels, both, width, label = 'Both')
ax.bar(labels, total, width, label = 'Total')
ax.set_ylabel('Count')
ax.set_xlabel('Platform Type')
ax.set_title('Breakdown of Xbox Game Pass Library by Platform Availability, 08/27/2020')
ax.legend()
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    if height == 0:
        continue
    x, y = p.get_xy() 
    ax.text(x+width/2, 
        y+height/2, 
        '{}'.format(height), 
        horizontalalignment='center', 
        verticalalignment='center')
plt.savefig('Xbox Game Pass Platform Breakdown.png')
```

Owning or having access to both a Xbox One and PC allows for unrestricted access to the full Xbox Game Pass library. In the unfortunate situation of singular platform access, Xbox-only individuals would have access to 14% (42) more games than their PC-only counterparts. These individuals would be able to play 83% (246) and 69% (204) of the game catalog, respectively.

### Microsoft Store price analysis

All the prices in 'df_game_list_clean' were pulled from Microsoft Store listings. This section compiles some key price stats for all games included, and also attempts to compare the PC only catalog's total value with what it would cost to outright purchase all games.

```python
# Microsoft Store list price stats
list_prices = df_game_list_clean.ListPrice.value_counts()
print(f'MS unique price count: {len(list_prices)}')
print(f'MS mean price: {round(df_game_list_clean.ListPrice.mean(), 2)}')
print('MS top 4 prices, counts: {}, {}, {}, {}'.format(*list_prices.items()))
microsoft_all_sum = df_game_list_clean.ListPrice.sum()
print(f'MS total value: {microsoft_all_sum}')
microsoft_pc_sum = round(df_game_list_clean.ListPrice[df_game_list_clean.Platform.isin(['PC', 'Both'])].sum(), 2)
print('MS total value, PC: {}'.format(microsoft_pc_sum))
microsoft_xbox_sum = round(df_game_list_clean.ListPrice[df_game_list_clean.Platform.isin(['Xbox One', 'Both'])].sum(), 2)
print('MS total value, Xbox: {}'.format(microsoft_xbox_sum))
```

The MS total values for Xbox and PC do not add up to overall MS total value. The single platform total values represent the cumulative value of games accessible on solely that platform, so there will be some overlap between the Xbox-only and PC-only tags (some games are offered on both platforms simultaneously).

A cursory inspection of this pricing information would appear to highlight a rather high bulk value to cost ratio for Xbox Game Pass. Rough calculations would put it at 532 (PC only) and 652 (Xbox only) months of sustained subscriptions to match the cost of purchasing all included games today. Combined, it would take just over 675 months of subscriptions to equal the cost of purchasing the entire library of games.

```python
# plot prices as histogram
from matplotlib import colors

fig, ax = plt.subplots(figsize = (20, 10))
N, bins, patches = ax.hist(df_game_list_clean.ListPrice, bins = 30)
fracs = N / N.max()
norm = colors.Normalize(fracs.min(), fracs.max())
for thisfrac, thispatch in zip(fracs, patches):
    color = plt.cm.viridis(norm(thisfrac))
    thispatch.set_facecolor(color)
ax.set_ylabel('Prices in USD')
ax.set_xlabel('Count')
ax.set_title('Xbox Game Pass Price Histogram')
plt.savefig('Xbox Game Pass Price Plot.png')
```

Far and away, the most common Microsoft Store game value in Xbox Game Pass is right around \$20. Outside of that, the game prices look to group around the \$5 and \$10 multiples. 

### Price comparison to outside game stores

I mentioned a cursory inspection because digging any deeper into Microsoft Store's own pricing reveals a rather large disparity between itself and other competing game marketplaces. Using Microsoft Store prices to gauge the total value of the Xbox Game Pass catalog is likely overinflating its true value. A more fair metric would be to use the lowest historical prices for games across all other stores, as this is far more likely to represent the decision a potential consumer would be facing: a subscription service for a slowly revolving assortment of games, or individual purchases of the same games whenever they are heavily discounted. Using historical lowest ever prices for game prices outside of the Microsoft Store represents the absolute lowest dollar sum that a customer could have paid to have the same game selection (a best case scenario for a non-Xbox Game Pass customer). The above mentioned historical PC game prices will be pulled from isthereanydeal.com.

A couple things to note, first that this section and onwards will only cover PC games as it vastly simplifies analysis. Next, local running of this notebook requires installation of the following things: Chrome / Chromium, Chromedriver, and Selenium. These are used to aid in web-scraping, specifically queries within a webpage and locating elements more easily than allowed through BeautifulSoup. Finally, you will need to substitute in your own 'chromium_driver_path' in the below cell, your path will likely not match mine. Onwards!

```python
# web scraping historical low prices from isthereanydeal.com, with the help of selenium
!pip install selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

chromium_driver_path = '/ProgramData/chocolatey/lib/chromedriver/tools/chromedriver'
driver = webdriver.Chrome(executable_path = chromium_driver_path)
driver.get('https://isthereanydeal.com/')
search_box = driver.find_element_by_xpath('//input[@id = "searchbox"]')
search_button = driver.find_element_by_xpath('//input[@type = "submit"]')
game_title = df_game_list_clean.ProductName[0]
search_box.send_keys(game_title)
search_button.click()
item_title = driver.find_element_by_xpath('//a[@class = "card__title"]').text
historical_low = driver.find_element_by_xpath('//div[@class = "numtag__primary"]').text
print(f'Title: {item_title}, Historical Low: {historical_low}')
driver.close()
```

In the above cell, the first game title from df_game_list_clean was queried on isthereanydeal.com. The printed result is the title of isthereanydeal.com's first query result along with the historical low value, in this case 'A Plague Tale: Innocence' and '$14.99'. It should be noted that since this is a simple query, it may return first results that do not match the desired game title. Manual clean-up of these conflicts will be required.

```python
# function that replicates the above cell, but in a headless fashion
def is_there_any_deal(game_title):
    options = Options()
    options.headless = True
    options.add_argument('--window-size=1920,1080')
    driver = webdriver.Chrome(options = options, executable_path = chromium_driver_path)
    driver.get('https://isthereanydeal.com/')
    search_box = driver.find_element_by_xpath('//input[@id = "searchbox"]')
    search_button = driver.find_element_by_xpath('//input[@type = "submit"]')
    search_box.send_keys(game_title)
    search_button.click()
    first_result = driver.find_element_by_xpath('//a[@class = "card__title"]').text
    historical_low = driver.find_element_by_xpath('//div[@class = "numtag__primary"]').text
    driver.close()
    return (first_result, historical_low)
```

```python
# a quick test to confirm it works
output = is_there_any_deal(game_title)
print(f'Title: {output[0]}, Historical Low: {output[1]}')
```

### Developer, publisher, and game genre trends

This final section of the notebook will take a look at trends between games, particularly who developed / published them and their unique characteristics (genre, features, rating, et cetera). All of this game data will come directly from Microsoft Store game webpages, links that were scraped way back in the first section of this notebook.

```python
# new dataframe for game details
titles = df_game_list_clean.ProductName[df_game_list_clean.Platform.isin(['PC', 'Both'])].values
links = df_game_list_clean.MSStoreLink[df_game_list_clean.Platform.isin(['PC', 'Both'])].values
df_game_details = pd.DataFrame(data = {'Title': titles, 'Link': links})
```

```python
# scraping additional game information from links collected in 'Collecting the list of games' section
chromium_driver_path = '/ProgramData/chocolatey/lib/chromedriver/tools/chromedriver'
driver = webdriver.Chrome(executable_path = chromium_driver_path)
driver.get(df_game_details.Link[0])
category_elements = driver.find_elements_by_xpath('//div[@id="category"]/div[1]/a')
print(f'Title: {df_game_details.Title[0]}')
print('Categories:',', '.join([str(item.get_attribute('aria-label'))[22:] for item in category_elements]))
driver.close()
```

```python
# function to repeat the above cell
def pull_details(row):
    driver.get(df_game_details.Link[row])
    category_elements = driver.find_elements_by_xpath('//div[@id="category"]/div[1]/a')
    categories = ', '.join([str(item.get_attribute('aria-label'))[22:] for item in category_elements])
    try:
        publishers = driver.find_element_by_xpath('//h4[text()="Published by"]//following-sibling::span').text.strip()
    except:
        publishers = ''
    try:
        developers = driver.find_element_by_xpath('//h4[text()="Published by"]//following-sibling::span').text.strip()
    except:
        developers = ''
    return categories, publishers, developers
```
